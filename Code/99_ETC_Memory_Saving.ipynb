```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from pyarrow import csv
```

```python
train = csv.read_csv('train_ver2.csv').to_pandas()
test = csv.read_csv('test_ver2.csv').to_pandas()
```

```python
train_and_test = [train, test]
```

```python
train.head()
```

```python
train.info()
```

```python
test.info()
```

```python
print(train.shape, test.shape)
```

```python
for dataset in train_and_test:
    only_int64 = dataset.select_dtypes(include= 'int64')
    print(only_int64.max())
```

```python
train_only_int64 = train.select_dtypes(include= 'int64')
test_only_int64 = test.select_dtypes(include = 'int64')

train_only_int64 = train_only_int64.drop('ncodpers', axis = 1)
test_only_int64 = test_only_int64.drop('antiguedad', axis = 1)

for i in train_only_int64.columns:
    train[i] = train[i].astype(np.int8)

for i in test_only_int64.columns:
    test[i] = test[i].astype(np.int8)
```

```python
train_only_float64 = train.select_dtypes(include= 'float64')
test_only_float64 = test.select_dtypes(include = 'float64')

print(train_only_float64.max())
print(test_only_float64.max())
```

```python
train_only_float64 = train_only_float64.drop('renta', axis = 1)

for i in train_only_float64.columns:
    train[i] = train[i].astype(np.float16)

for i in test_only_float64.columns:
    test[i] = test[i].astype(np.float16)
```

```python
for i in only_float64:
    print(only_float64[i].value_counts())
```
